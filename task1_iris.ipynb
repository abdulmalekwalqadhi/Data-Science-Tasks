## Problem Statement
The goal is to explore the Iris dataset and identify relationships between different features using visualizations.

## Dataset Description

The Iris dataset contains measurements of iris flowers.

Features:
- Sepal Length
- Sepal Width
- Petal Length
- Petal Width

Target:
- Species (Setosa, Versicolor, Virginica)

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
df.shape
df.columns
df.head()
```

## Exploratory Data Analysis

We visualize relationships between variables.

```
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Scatter Plot")
plt.show()

df['sepal_length'].hist()
plt.title("Histogram")
plt.show()

sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Box Plot")
plt.show()
```
